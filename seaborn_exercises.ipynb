# 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
chipotle = pd.read_csv('chipotle.csv')

chipotle.head()
```

```python
chipotle_quantity = chipotle.groupby('item_name').quantity.sum()
# Create list of items and quantity of item sold

chipotle_quantity.head()
```

```python
revenue = chipotle.groupby('item_name').item_price.agg('sum')
# Create list of items and value of money each item earned

revenue.head()
```

```python
chipotle_quantity_and_revenue = pd.merge(chipotle_quantity, revenue,left_on='item_name',right_on='item_name', how='left')
# Combine lists to create a dataframe of quantity of items sold and amount of money earned 

chipotle_quantity_and_revenue.head()
```

```python
top_four_revenue = chipotle_quantity_and_revenue.sort_values(['quantity', 'item_price'], ascending=[False, False]).head(4)
# Condense dataframe so that only the top 4 items that sold the most and made the most money remain

top_four_revenue
```

```python
ax = sns.barplot(x = top_four_revenue.index, y = top_four_revenue.item_price)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

# 5. Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
sleep = data('sleepstudy')

sleep.head()
```

```python
sleep['Subject'] = 'Subject_' + sleep.Subject.astype(str)
# Convert subject into catagorical value

sleep.Subject.unique()
```

```python
avg_reaction = sleep.groupby('Days').Reaction.mean()
# Create an average reaction time array

avg_reaction
```

```python
plt.figure(figsize = (16,9))

sns.lineplot(data = sleep, x='Days', y='Reaction', hue='Subject')

sns.lineplot(data = avg_reaction, linewidth = 5 , color = 'black', estimator = 'mean')

plt.legend()

plt.show()
```
