# 1. Use matplotlib to plot the following equation:

## y=x**2−x+2

```python
x = range(-10,10)

y = [n ** 2 - n + 2 for n in x]

plt.plot(x,y)

plt.title('y = (x * x) − x + 2')

plt.annotate('The Origin', xy = (0, 0), xytext = (2, 38), arrowprops = dict(facecolor = 'gray'))

plt.show()
```

# 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(14, 6))
## make figure size

plt.subplot(2,2,1)
x = range(0,50)
y = [math.sqrt(n) for n in x]
plt.title('y = sqrt.x')
plt.plot(x,y)
## make first subplot
### y=sqrt.x

plt.subplot(2,2,2)
x = range(-10,10)
y = [n ** 3 for n in x]
plt.title('y = x ** 3')
plt.plot(x,y)
## make second subplot
### y=x**3

plt.subplot(2,2,3)
x = range(-10,10)
y = [2 ** n for n in x]
plt.title('y = 2 ** x')
plt.plot(x,y)
## make thrid subplot
### y=2**x

plt.subplot(2,2,4)
x = range(1,10)
y = [1 / (n + 1) for n in x]
plt.plot(x,y)
plt.title('y = 1 / (x + 1)')
## make fourth subplot
### y=1/(x+1)

plt.tight_layout()
## make plots look neat
plt.show()
```

```python
plt.figure(figsize=(14, 6))
## make figure size


x1 = range(0,10)
y1 = [math.sqrt(n) for n in x1]

plt.plot(x1, y1, color='red', label='sqrt.x')
## make first line
### y=sqrt.x


x2 = range(-10,10)
y2 = [n ** 3 for n in x2]

plt.plot(x2, y2, color='blue', label='x**3')
## make second line
### y=x**3


x3 = range(-10,10)
y3 = [2 ** n for n in x3]

plt.plot(x3 ,y3, color='green', label='2**x')
## make thrid line
### y=2**x


x4 = range(1,10)
y4 = [1 / (n + 1) for n in x4]
plt.plot(x4 ,y4, color='yellow', label='1/(x+1)')

## make fourth line
### y=1/(x+1)

plt.title('Y =')
plt.legend()
plt.tight_layout()
plt.ylim([-100, 100])
## make plots look neat
plt.show()
```

### Curves to graph

y=0n+1 and label the curve "O(1)"

y=n! and label it "O(n!)"

y=nn and label it "O(n^n)"

```python

plt.figure(figsize = (15, 10))


x1 = range(1,100)
x2 = range(1,100)
x3 = range(1,100)
x4 = range(1,100)
x5 = range(1,100)
x6 = range(1,100)
x7 = range(1,100)
x8 = range(1,100)

y1 = [(0 * n) + 1 for n in x1]
y2 = [math.log(n) for n in x2]
y3 = x3
y4 = [n * (math.log(n)) for n in x4]
y5 = [n ** 2 for n in x5]
y6 = [2 ** n for n in x6]
y7 = [math.factorial(n) for n in x7]
y8 = [n ** n for n in x8]

plt.plot(x1, y1, label='O(1)')
plt.plot(x2, y2, label='O(log n)')
plt.plot(x3, y3, label='O(n)')
plt.plot(x4, x4, label='(n log n)')
plt.plot(x5, y5, label='O(n^2)')
plt.plot(x6, y6, label='O(2^n)')
plt.plot(x7, y7, label='O(n!)')
plt.plot(x8, y8, label='(n^n)')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.ylim(0,50)

plt.legend()

plt.show()
```
